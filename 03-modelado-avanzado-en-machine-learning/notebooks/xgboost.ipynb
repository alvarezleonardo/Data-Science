# XGBoost sobre California Housing

**XGBoost** (eXtreme Gradient Boosting) es una implementacion optimizada de gradient boosting: usa gradientes de segundo orden, es rapida, paraleliza y **maneja valores nulos automaticamente**.

Dataset `housing` (California): predecimos `median_house_value` (valor mediano de la vivienda) a partir de variables geograficas y demograficas.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from xgboost import XGBRegressor
```

## Datos

`ocean_proximity` es categorica → la convertimos a variables dummy. `total_bedrooms` tiene nulos, pero **no** los imputamos: XGBoost aprende solo la mejor direccion para las divisiones con datos faltantes.

```python
housing = pd.read_csv("../datasets/housing.csv")
housing.head()
```

```python
housing = pd.get_dummies(housing, columns=['ocean_proximity'], drop_first=True)

x = housing.drop("median_house_value", axis=1)
y = housing["median_house_value"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=127)
```

## Modelo

Evaluamos con R2 (`reg_score`), MSE y RMSE. El RMSE se muestra multiplicado por 100.

```python
model=XGBRegressor(n_estimators=30)
model.fit(x_train,y_train)
preds=model.predict(x_test)

reg_score=r2_score(y_test,preds)
print(reg_score)

mse=mean_squared_error(y_test,preds)
print(mse)

rmse=mse**0.5
print(rmse*100)
```

Con solo 30 arboles, XGBoost alcanza un R2 de ~0.83 sobre test (explica el 83% de la varianza del valor de las viviendas). Subir `n_estimators` y ajustar `learning_rate`/`max_depth` suele mejorar el resultado.
