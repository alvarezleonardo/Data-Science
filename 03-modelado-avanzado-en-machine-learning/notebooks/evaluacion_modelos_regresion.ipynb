# Evaluacion de modelos de regresion (scikit-learn)

Regresion lineal con scikit-learn sobre el dataset `bikes`: ajuste con una y varias variables, y evaluacion del error con metricas (MAE, MSE, RMSE, R2) y train/test split.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression

bikes = pd.read_csv("../datasets/bikes.csv", index_col="datetime", parse_dates=True)
bikes.shape
```

```python
bikes.sample(10)
```

## Regresion lineal simple

Una sola variable predictora (`temp`) para estimar `count`.

```python
feature_cols = ["temp"]
X = bikes[feature_cols]
y = bikes["count"]

linreg = LinearRegression()
linreg.fit(X, y)

print("Intercepto:  ", linreg.intercept_)
print("Coeficientes:", linreg.coef_)
```

```python
# Prediccion para temp = 25: a mano y con sklearn
temp = 25
print("A mano: ", linreg.intercept_ + linreg.coef_ * temp)

X_new = pd.DataFrame({"temp": [temp]})
print("sklearn:", linreg.predict(X_new))
```

## Regresion lineal multiple

Varias variables predictoras. Antes de ajustar miramos la correlacion entre los predictores.

```python
feature_cols = ["temp", "weather", "humidity"]
bikes[feature_cols].corr()
```

```python
sns.heatmap(bikes[feature_cols].corr(), vmin=-1, vmax=1, center=0, cmap="YlGnBu");
```

```python
X = bikes[feature_cols]
y = bikes["count"]

linreg = LinearRegression()
linreg.fit(X, y)

print("Intercepto:", linreg.intercept_)
list(zip(feature_cols, linreg.coef_))
```

## Metricas de evaluacion

MAE, MSE, RMSE y R2 sobre un ejemplo de valores reales vs predichos.

```python
from sklearn import metrics

true = [10, 7, 5, 5, 10, 8, 8, 15, 12]
pred = [12, 8, 3, 4, 12, 9, 8, 12, 13]

print("MAE: ", metrics.mean_absolute_error(true, pred))
print("MSE: ", metrics.mean_squared_error(true, pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(true, pred)))
print("R2:  ", metrics.r2_score(true, pred))
```

## Evaluacion con train/test split

Entrenamos en una parte de los datos y medimos el error sobre datos no vistos. Usamos `shuffle=False` porque `bikes` es una serie temporal.

```python
from sklearn.model_selection import train_test_split

def train_test_error(feature_cols):
    X = bikes[feature_cols]
    y = bikes["count"]
    X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    y_pred = linreg.predict(X_test)

    print(feature_cols)
    print("y_test sample:", y_test.values[0:10])
    print("y_pred sample:", y_pred[0:10].astype(int))
    print("MAE: ", metrics.mean_absolute_error(y_test, y_pred))
    print("MSE: ", metrics.mean_squared_error(y_test, y_pred))
    print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print("R2:  ", metrics.r2_score(y_test, y_pred))
    print()
```

```python
train_test_error(["temp", "humidity"])
train_test_error(["temp", "humidity", "windspeed"])
```
